## requests + BeautifulSoup


```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.dcard.tw/f'

r = requests.get(url)
r.encoding = 'utf-8'
print(r.text[0:3000])
```

```python
print(type(r.text))
```

```python
soup = BeautifulSoup(r.text)
# print(type(soup.text))
print(soup)
```

## Grab + PyQuery

```python
from grab import Grab
g = Grab()
resp = g.go('https://www.dcard.tw/f')
resp.body
```

```python
print(type(resp.body))
```

```python
from pyquery import PyQuery as pq

doc = pq(resp.body)
print(doc)
```

```python
print(type(doc))
```
